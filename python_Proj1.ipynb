```python
class Student:
    def __init__(self, name, roll_no, marks=None):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks if marks else []

    def add_marks(self, marks):
        self.marks = marks

    def total_marks(self):
        return sum(self.marks)

    def average_marks(self):
        return sum(self.marks) / len(self.marks) if self.marks else 0

    def __str__(self):
        return f"Name: {self.name}, Roll No: {self.roll_no}, Marks: {self.marks}, Total: {self.total_marks()}, Average: {self.average_marks()}"


class StudentRecords:
    def __init__(self):
        self.records = {}

    def add_student(self, name, roll_no, marks):
        if roll_no not in self.records:
            self.records[roll_no] = Student(name, roll_no, marks)
            print(f"Student {name} added successfully.")
        else:
            print(f"Student with Roll No {roll_no} already exists.")

    def update_marks(self, roll_no, marks):
        if roll_no in self.records:
            self.records[roll_no].add_marks(marks)
            print(f"Marks updated for Roll No {roll_no}.")
        else:
            print(f"Student with Roll No {roll_no} does not exist.")

    def find_topper(self):
        if not self.records:
            print("No students records found.")
            return
        topper = max(self.records.values(), key=lambda student: student.total_marks())
        print(f"Topper: {topper.name} (Roll No: {topper.roll_no}) with Total Marks: {topper.total_marks()}")

    def display_all_records(self):
        if not self.records:
            print("No student records available.")
            return
        for student in self.records.values():
            print(student)


# Main program to interact with the user
def main():
    records = StudentRecords()

    while True:
        print("\nStudent Record Management System")
        print("1. Add Student Details")
        print("2. Update Marks")
        print("3. Find Topper")
        print("4. Display All Records")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            name = input("Enter student's name: ")
            roll_no = input("Enter student's roll number: ")
            marks = list(map(int, input("Enter marks (separated by spaces): ").split()))
            records.add_student(name, roll_no, marks)

        elif choice == '2':
            roll_no = input("Enter roll number of student to update marks: ")
            marks = list(map(int, input("Enter new marks (separated by spaces): ").split()))
            records.update_marks(roll_no, marks)

        elif choice == '3':
            records.find_topper()

        elif choice == '4':
            records.display_all_records()

        elif choice == '5':
            print("Exiting...")
            break

        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    main()
```
